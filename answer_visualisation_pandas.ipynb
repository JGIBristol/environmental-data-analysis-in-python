```python
import pandas as pd

temperature = pd.read_csv(
    "https://milliams.com/courses/data_analysis_python/cetml1659on.txt",  # file name
    skiprows=6,  # skip first 6 rows of the header
    delim_whitespace=True,  # whitespace separated columns
    na_values=["-99.9", "-99.99"],  # NaNs
)
```

```python
year_plot = temperature["JAN"].plot(xlabel="Year", ylabel=r"Temperature ($\degree$C)")
```
